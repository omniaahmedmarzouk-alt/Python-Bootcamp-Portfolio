 **Task 1**: Print numbers 0 to 2 using while and print 'Loop finished.'

```python
n=0
while n<3:
    print(n)
    n+=1
print('Loop finished.')
```

### **Task 2**: Print prime numbers between 5 and 100.

```python
n=5
while n<100:
    m=2
    prime=True
    while n>m:
        if(n%m==0):
          prime=False
          break
        m+=1
    if prime:
      print(n)
    n+=1
```

## **Task 3**: Generate first 15 Fibonacci numbers.





```python
a=0
b=1
n=0
while n<15:
    print (a)
    new=a+b
    a=b
    b=new
    n+=1
```

### Task 4: ATM simulation with withdrawal until balance=0.

```python
balance =1000
while balance>0:
    print(f"Current Balance: {balance}")
    withdraw=int(input('Enter withdraw: '))
    if(withdraw<=balance):
       balance-=withdraw
       print(f"Remaining: {balance}")
    else:
        print('Enter valid number.')
print('balance became 0.')
```

### **Task 5**: Find common elements between two lists.

```python
l1=[5,40,6,8]
l2=[4,6,40,7]
list_=[]
for i in l1:
    if i in l2:
        list_.append(i)
print(list_)
```

### **Task 6**: Compute factorial of 10.

```python
n=10
f=1
while n>0:
    f*=n
    n-=1
print(f)
```

### **Task 7**: Ask user for number >7 with validation.





```python
while True:
    n=int(input('Enter number: '))
    if(n>7):
        print("Valid number!")
        break
    else:
        print("Wrong input. Number must be > 7.")
```

### **Task 8**: Sum digits of 12347843.

```python
num=12347843
str_num=str(num)
sum=0
for i in str_num:
    sum+=int(i)
print(sum)
```

 **Task 9**: Check if 'level' is palindrome.

```python
word = 'level'
if word == word[::-1]:
    print("It is a palindrome")
else:
    print("Not a palindrome")
```

### **Task 10**: Create dictionary from two lists.

```python
keys = ['Name', 'Age', 'Major']
values = ['Omnia', 21, 'AI Engineer']
my_dict = {}

for i in range(len(keys)):
    key_item = keys[i]      
    value_item = values[i]  
    my_dict[key_item] = value_item

print(my_dict)
```
