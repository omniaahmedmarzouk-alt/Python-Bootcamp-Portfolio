## Task 1: Sum of Even Numbers
Write a function that takes a list of numbers and returns the sum of only the even numbers.

Test list:
[1, 2, 3, 4, 5, 6]

```python
# Write your solution here
def Even(lst):
    sum=0
    for i in lst:
        if (i%2==0): sum+=i
    return sum
print(Even([1,2,3,4,5]))
```

## Task 2: Count Words in a Sentence
Write a function that takes a sentence as input and returns the number of words.

Test sentence:
"Python makes programming easy"

```python
# Write your solution here
def count_Word(Word):
   Words=Word.split()
   return len(Words)

print(count_Word('Python makes programming easy'))
```

## Task 3: Find the Smallest Number
Write a function that takes a list of numbers and returns the smallest number.

Test list:
[45, 12, 78, 3, 25]

```python
# Write your solution here
def Smallest_num(lst):
    return min(lst)

print(Smallest_num( [45, 12, 78, 3, 25]))
```

## Task 4: Reverse a String
Write a function that takes a string and returns the reversed version.

Test word:
"education"

```python
# Write your solution here
def reversed(word):
  return word[::-1]

print(reversed('education'))
```

## Task 5: Check Positive or Negative
Write a function that takes a number and prints whether it is positive, negative, or zero.

Test values:
10, -5, 0

```python
# Write your solution here
def check_sign(num):
  if(num>0):    print('positive')
  elif(num==0): print('zero')
  else:         print('negative')

check_sign(10)
check_sign(-5)
check_sign(0)
```

## Task 6: Count Vowels in a String
Write a function that takes a string and returns the number of vowels.

Test text:
"Artificial Intelligence"

```python
# Write your solution here
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

print(count_vowels("Artificial Intelligence"))
```
