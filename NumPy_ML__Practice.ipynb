# NumPy for Machine Learning & Job Market



## Task 1: Data Normalization

```python

import numpy as np
features = np.array([50, 60, 70, 80, 90])

# TODO: Calculate mean
Mean=np.mean(features)
# TODO: Calculate standard deviation
Std=np.std(features)
# TODO: Normalize features
normalized_features = (features - Mean) / Std

print("Mean:", Mean)
print("Std:", Std)
print("Normalized:", normalized_features)
```

## Task 2: Feature Selection

```python

data = np.array([0.2, 0.8, 0.5, 1.2, 0.3])
threshold = 0.6

# TODO: Select values above threshold
Selected = data[data>threshold]
print("Selected = ", Selected)
```

## Task 3: Distance Calculation

```python

user1 = np.array([3, 4, 5])
user2 = np.array([1, 2, 3])

# TODO: Compute Euclidean distance
euclidean_distance=np.linalg.norm(user1 - user2)
print("Euclidean= ", euclidean_distance)
```

## Task 4: Dataset Statistics

```python

dataset = np.array([12, 15, 18, 20, 22, 25])

# TODO: Compute mean
Mean=np.mean(dataset)
# TODO: Compute std
Std=np.std(dataset)
# TODO: Compute min and max
Max=np.max(dataset)
Min=np.min(dataset)

print(f"Mean: {Mean}, Std: {Std}")
print(f"Min: {Min}, Max: {Max}")
```

## Task 5: Outlier Detection

```python

salaries = np.array([3000, 3200, 3100, 9000, 3300])

# TODO: Compute mean and std
mean_sal = np.mean(salaries)
std_sal = np.std(salaries)

# TODO: Detect outliers
outliers = salaries[np.abs(salaries - mean_sal) > std_sal]

print("Mean Salary:", mean_sal)
print("Detected Outliers:", outliers)
```

## Task 6: Linear System

```python

A = np.array([[3,1],[1,2]])
b = np.array([9,8])

# TODO: Solve system Ax=b
x = np.linalg.solve(A, b)

print("Solution x:", x)
```

## Task 7: Correlation

```python

feature1 = np.array([1,2,3,4,5])
feature2 = np.array([2,4,6,8,10])

# TODO: Compute correlation coefficient
correlation_matrix = np.corrcoef(feature1, feature2)
correlation = correlation_matrix[0, 1]

print("Correlation Matrix:\n", correlation_matrix)
print("Correlation Coefficient:", correlation)
```

## Task 8: Random Simulation

```python

# TODO: Generate 100 samples from normal distribution
Samples=np.random.normal(loc=0,scale=1,size=100)

# TODO: Compute mean and std
Mean=np.mean(Samples)
Std=np.std(Samples)

print(f"Generated Mean: {Mean}")
print(f"Generated Std: {Std}")
```

## Task 9: Matrix Rank

```python

features = np.array([[1,2,3],[2,4,6],[3,6,9]])

# TODO: Compute matrix rank
rank = np.linalg.matrix_rank(features)

print("Matrix Rank:", rank)
```

## Task 10: Demand Forecast

```python

demand = np.array([120,130,125,140,135,150])

# TODO: Compute mean and std
mean_demand = np.mean(demand)
std_demand = np.std(demand)

# TODO: Simulate next month demand
next_month = np.random.normal(mean_demand, std_demand)

# TODO: Round results
forecast_rounded = np.round(next_month)

print(f"Historical Mean: {mean_demand}")
print(f"Forecast for next month: {forecast_rounded}")
```
