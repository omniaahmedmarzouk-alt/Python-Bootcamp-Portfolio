## Task 1: Student Attendance System
Create a class to track student attendance. Methods: mark_attendance, get_attendance.


```python
class AttendanceSystem:
    def __init__(self):
        self.attendance = {}

    def mark_attendance(self, student_name):
        if student_name in self.attendance:
            self.attendance[student_name] += 1
        else:
            self.attendance[student_name] = 1

        print(f"Attendance marked for: {student_name}")

    def get_attendance(self, student_name):
        return self.attendance.get(student_name, 0)


school = AttendanceSystem()

school.mark_attendance("Ahmed")
school.mark_attendance("Ahmed")

school.mark_attendance("Omar")

print(f"Ahmed attended: {school.get_attendance('Ahmed')}")
print(f"Omar attended: {school.get_attendance('Omar')}")
```

## Task 2: Online Shopping Cart
Create a class that calculates total cart price including discounts or taxes.


```python
class Shopping:
    def __init__(self):
        self.items = []

    def add_items(self, product, price):
        self.items.append((product, price))
        print(f"Item '{product}' Added.")

    def calculate_price(self, discount):
        total = 0
        for item in self.items:
            price = item[1]
            total += price
        final_price = total - discount
        return final_price

cart = Shopping()
cart.add_items("Laptop", 15000)
cart.add_items("Mouse", 500)

print(f"Total after discount: {cart.calculate_price(100)}")
```

## Task 3: Employee Attendance Percentage
Create a class that calculates attendance percentage for each employee.



```python
class Employee:
    def __init__(self, name, total_days, attended_days):
        self.name = name
        self.total_days = total_days
        self.attended_days = attended_days

    def check_percentage(self):
        return (self.attended_days / self.total_days) * 100


employee1 = Employee('Ahmed', 20, 15)
employee2 = Employee('Adham', 25, 20)

print(f"Percentage for {employee1.name}: {employee1.check_percentage()}%")
print(f"Percentage for {employee2.name}: {employee2.check_percentage()}%")
```

## Task 4: Course Enrollment System
Create a class that tracks enrolled students and the courses they join.


```python
class CourseSystem:
    def __init__(self):
        self.enrollments = {}

    def enroll(self, st_name, st_courses):
        if st_name in self.enrollments:
            self.enrollments[st_name].extend(st_courses)
        else:
            self.enrollments[st_name] = st_courses
            print(f"Student {st_name} enrolled successfully.")

    def show_data(self, st_name):
        return self.enrollments.get(st_name, "Student not found")

my_system = CourseSystem()
my_system.enroll('Ahmed', ['AI', 'ML', 'DB'])
my_system.enroll('Esam', ['Math', 'Physics'])

print(f"Ahmed has courses: {my_system.show_data('Ahmed')}")
print(f"Esam has courses: {my_system.show_data('Esam')}")
```

## Task 5: Hospital Patient Visits
Create a class to count patient visits and print daily reports.


```python
class Hospital:
  def __init__(self):
    self.visits={}

  def add_visit(self,patient_name):
    if patient_name in self.visits:
       self.visits[patient_name]+=1
    else:
       self.visits[patient_name]=1

  def print_report(self):
     for name, count in self.visits.items():
       print(f"Patient {name}: {count} visits")

system=Hospital()
system.add_visit('Omar')
system.add_visit('Omar')
system.add_visit('Ahmed')
system.add_visit('Omar')
system.add_visit('Amr')
system.print_report()
```

## Task 7 (ML): Model Accuracy Tracker
Create a class to calculate ML model accuracy based on correct predictions and total predictions.



```python
class ModelAccuracy:
    def __init__(self):
        self.total_correct = 0
        self.total_samples = 0

    def update_metrics(self, correct, total):
        self.total_correct += correct
        self.total_samples += total

    def get_accuracy(self):
        if self.total_samples == 0:
            return 0.0
        return self.total_correct / self.total_samples *100

model = ModelAccuracy()

model.update_metrics(80, 100)
print(f"Accuracy after batch 1: {model.get_accuracy()}%")
model.update_metrics(90, 100)
print(f"Accuracy after batch 2: {model.get_accuracy()}%")
```

## Task 8: Electricity Meter
Create a class to calculate electricity bill based on usage and rate per kWh.




```python
class ElectricityMeter:
    def __init__(self, rate_per_kwh):
        self.rate = rate_per_kwh
        self.total_usage = 0

    def add_usage(self, kwh):
        self.total_usage += kwh

    def calculate_bill(self):
        return self.total_usage * self.rate

meter = ElectricityMeter(1.5)
meter.add_usage(100)
meter.add_usage(200)
print(f"Total Bill: {meter.calculate_bill()}")
```

## Task 9 (ML): Prediction Threshold
Create a class to apply a probability threshold to ML model predictions (e.g., 0.5).



```python
class ModelThreshold:
  def __init__(self,threshold):
    self.threshold=threshold

  def apply_threshold(self,probability):
    if(probability>=self.threshold):
      return 1
    return 0

predict=ModelThreshold(0.7)
print(f"the probability of 0.8 is: {predict.apply_threshold(0.8)}")
print(f"the probability of 0.4 is: {predict.apply_threshold(0.4)}")
```

## Task 10 (ML): Loss Function (MAE)
Create a class to calculate Mean Absolute Error between predicted and actual values.

```python
class MAE:
    def __init__(self):
        self.total_error = 0
        self.count = 0

    def add_prediction(self, predicted, actual):
        diff = predicted - actual
        error = abs(diff)
        self.total_error += error
        self.count += 1

    def calculate_mae(self):
        if self.count == 0: return 0
        return self.total_error / self.count

loss = MAE()
loss.add_prediction(10, 12)
loss.add_prediction(5, 5)
loss.add_prediction(8, 4)

print(f"Mean Absolute Error (MAE): {loss.calculate_mae()}")
```
